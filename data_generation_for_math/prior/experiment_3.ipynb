```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_a = [
    {"x": 0.0, "y": 14.0},
    {"x": 1.0, "y": 6.3},
    {"x": 2.0, "y": 2.835},
    {"x": 3.0, "y": 1.276},
    {"x": 4.0, "y": 0.574},
    {"x": 5.0, "y": 0.258},
    {"x": 6.0, "y": 0.116},
    {"x": 7.0, "y": 0.052},
    {"x": 8.0, "y": 0.024},
    {"x": 9.0, "y": 0.011},
    {"x": 10.0, "y": 0.005}
]

dataset_b = [
    {"x": 0.0, "y": 13.0},
    {"x": 0.5, "y": 10.625},
    {"x": 1.0, "y": 8.5},
    {"x": 1.5, "y": 6.625},
    {"x": 2.0, "y": 5.0},
    {"x": 2.5, "y": 3.625},
    {"x": 3.0, "y": 2.5},
    {"x": 3.5, "y": 1.625},
    {"x": 4.0, "y": 1.0},
    {"x": 4.5, "y": 0.625},
    {"x": 5.0, "y": 0.5}
]
```

```python
def generate_new_dataset(dataset, x_noise_std_dev=0.1, y_noise_std_dev=0.1):
    target_dataset = []
    for point in dataset:
        x = point['x'] + np.random.normal(0, x_noise_std_dev)
        y = point['y'] + np.random.normal(0, y_noise_std_dev)
        target_dataset.append({'x': x, 'y': y})
    return target_dataset

target_dataset = generate_new_dataset(dataset_b)
```

```python
x1 = [point['x'] for point in dataset_a]
y1 = [point['y'] for point in dataset_a]

x2 = [point['x'] for point in dataset_b]
y2 = [point['y'] for point in dataset_b]

x3 = [point['x'] for point in target_dataset]
y3 = [point['y'] for point in target_dataset]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label='y=14*0.45^x')
plt.plot(x2, y2, label='y=0.5x^2-5x+13')
plt.plot(x3, y3, label='target')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the two functions')
plt.legend()
plt.show()
```

```python
data = [{'x': 0.20579380018140975, 'y': 12.994281554772888},
 {'x': 0.4127340603822577, 'y': 10.418881481048789},
 {'x': 0.7738114254986388, 'y': 8.392571049560841},
 {'x': 1.604436149603848, 'y': 6.490223845110359},
 {'x': 2.0111563678936424, 'y': 4.965206213585994},
 {'x': 2.4655274070834907, 'y': 3.6413317001365932},
 {'x': 3.11818390900918, 'y': 2.5781156803375964},
 {'x': 3.7223981339791403, 'y': 1.6315918192478691},
 {'x': 4.350651825120238, 'y': 0.9234787410613083},
 {'x': 4.701111083026894, 'y': 0.5543270637473905},
 {'x': 4.874676049584157, 'y': 0.34903872159951727}]

x_values = [item['x'] for item in data]
y_actual = [item['y'] for item in data]

# function definitions
def f_a(x):
    return 14*0.45**x

def f_b(x):
    return 0.5*x**2 - 5*x + 13

# predicted y values for each function
y_pred_a = [f_a(x) for x in x_values]
y_pred_b = [f_b(x) for x in x_values]

# RMSE calculations
rmse_a = np.sqrt(np.mean((np.array(y_actual) - np.array(y_pred_a))**2))
rmse_b = np.sqrt(np.mean((np.array(y_actual) - np.array(y_pred_b))**2))

print('RMSE for function a:', rmse_a)
print('RMSE for function b:', rmse_b)
```
