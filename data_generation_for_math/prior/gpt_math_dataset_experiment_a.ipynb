```python
import matplotlib.pyplot as plt

from functions import gpt
```

```python
model = 'gpt-4'
```

```python
system_prompt = """You are an expert mathmatician. 
- Please take the function provided and return a dataset that fits the function 
- Please format your response in JSON. You only speak JSON. Do not write text that isn't JSON.
- If you do not know the answer return {}
"""
```

```python
text = 'y=14*0.45^x'
```

```python
response = gpt.chat_create(system_prompt, text, model, output_json=True)
```

```python
response
```

```python
x_values = [item['x'] for item in response['dataset']]
y_values = [item['y'] for item in response['dataset']]
```

```python
plt.plot(x_values, y_values, marker='o')
plt.title("Data Plot")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
