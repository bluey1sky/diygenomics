```python
import numpy as np
import os
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
base_name = "2021_Hashimoto_Neural_ODE_and_holographic_QCD_PUB"
project_folder = "diygenomics-projects"
sub_category = "math"
work_bucket = "AdS-CFT"
input_file = 'extracted_annotated_math.csv'
index_col = 'uuid'
# external_id = "2023_05_22_92dc0613b4493d7b5847g"
```

```python
data_path = os.getenv('DATA_PATH')
file_path = lambda *args: os.path.join(data_path, project_folder, sub_category, work_bucket,
                                       base_name, 'mathpix', *args)
```

```python
df = pd.read_csv(file_path(input_file), index_col=index_col)
```

```python
def create_embedding(row):
    math = row['math']
    math = math.replace('\[', '').replace('\]', '').replace('\n', '')

    chars = sorted(list(set(math)))

    # Create a dictionary mapping characters to their indices
    char_to_idx = {char: idx for idx, char in enumerate(chars)}

    # Create one-hot embeddings
    embeddings = np.zeros((len(math), len(chars)))

    for i, char in enumerate(math):
        embeddings[i, char_to_idx[char]] = 1

    print(embeddings)
    print(len(embeddings))
```

```python
row = df.iloc[1]
create_embedding(row)
```
